```python
# important requriment library
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import f_classif
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt
```

```python
# select the kbest function call to rms or chi2
def selectkbest(indep_X,dep_Y,n):
        test = SelectKBest(score_func=f_classif, k=n)# k=n is call to the number
    # input variable and output variables
        fit1= test.fit(indep_X,dep_Y)
    #create a model and then return to the generated feature model
        selectk_features = fit1.transform(indep_X)
        return selectk_features
```

All models have the same column utilised for ex: logistics, SVM, KNN, and Naive Bayes. 
The selected model is determined by the split the Standard scaler functions used.

```python
def split_scalar(indep_X,dep_Y):
    X_train,X_test,y_train,y_test=train_test_split(indep_X,dep_Y, test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train)
    X_test=sc.fit_transform(X_test)
    return X_train,X_test,y_train,y_test
```

```python
def cm_prediction(classifier,X_test):
    y_pred=classifier.predict(X_test)
    # Making the Confusion Matrix
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    Accuracy=accuracy_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    return classifier,Accuracy,report,X_test,y_test,cm
    # Call to the split and standard scaler to create the one-by-one function created for the return 
    # statement x- test,y-test,and cm is  function is running
```

```python
# Aboue as  the cm Same to the all model runing
def logistic(X_train,y_train,X_test):
    # Fitting K-NN to the Training set
    from sklearn.linear_model import LogisticRegression
    classifier =LogisticRegression(random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def svm_linear(X_train,y_train,X_test):
    # Fitting K-NN to the Training set
    from sklearn.svm import SVC
    classifier =SVC(kernel = 'linear',random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def svm_NL(X_train,y_train,X_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='rbf', gamma='scale',random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def Navie(X_train,y_train,X_test):
    # Fitting K-NN to the Training set
    from sklearn.naive_bayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def knn(X_train,y_train,X_test):
    # Fitting K-NN to the Training set
    from sklearn.neighbors import KNeighborsClassifier
    classifier=KNeighborsClassifier(n_neighbors=7,weights='uniform',algorithm='auto',leaf_size=30,p=2,metric='minkowski')
    classifier.fit(X_train,y_train)
    classifier,Accuraacy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def DecisionTree(X_train,y_train,X_test):
    # Fitting K-NN to the Training set
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier(criterion='entropy', splitter='random',max_features='sqrt',random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
def RandomForest(X_train,y_train,X_test):
     # Fitting K-NN to the Training set
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier(n_estimators = 10, criterion = 'entropy',max_features='sqrt', random_state = 0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
    return classifier,Accuracy,report,X_test,y_test,cm
```

```python
# all model of the same table create are created to funcation
def selectk_Classification(acclog,accsvml,accsvmnl,accnav,accknn,accDecisionTree,accRandomForest):
    dataframe=pd.DataFrame(index=['ChiSquare'],columns=['Logistic','SVMl','SVMnl','Navie','knn','DecisionTree','RandomForest'])
    for number, idex in enumerate(dataframe.index):
        dataframe['Logistic'][idex]=acclog[number]
        dataframe['SVMl'][idex]=accsvml[number]
        dataframe['SVMnl'][idex]=accsvmnl[number]
        dataframe['Navie'][idex]=accnav[number]
        dataframe['knn'][idex]=accknn[number]
        dataframe['DecisionTree'][idex]=accDecisionTree[number]
        dataframe['RandomForest'][idex]=accRandomForest[number]
        return dataframe # return to the funcation to dataframe
```

```python
# upload the csv file, and check the for the null values.
dataset1=pd.read_csv('prep.csv',index_col=None)
df2=dataset1 # back up of the original dataset
df2=pd.get_dummies(df2,drop_first=True)
df2
```

```python
# independent and dependent variables suppurating to the funcation
indep_X = df2.drop(['classification_yes'], axis=1)
dep_Y = df2['classification_yes']
```

```python
# Call the k-best independent department x, y to create an accuracy list using the provided parameters and set k = n.
kbest=selectkbest(indep_X,dep_Y,7)
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]
```

```python
X_train,X_test,y_train,y_test=split_scalar(kbest,dep_Y)
#  The funcation call to the appends to the accuracy same to the another model
classification,Accuracy,report,X_test,y_test,cm=logistic(X_train,y_train,X_test)
acclog.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=svm_linear(X_train,y_train,X_test)
accsvml.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=svm_NL(X_train,y_train,X_test)
accsvmnl.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=Navie(X_train,y_train,X_test)
accnav.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=knn(X_train,y_train,X_test)
accknn.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=DecisionTree(X_train,y_train,X_test)
accdes.append(Accuracy)

classification,Accuracy,report,X_test,y_test,cm=RandomForest(X_train,y_train,X_test)
accrf.append(Accuracy)
# The All list input and create the table in the run function
result=selectk_Classification(acclog,accsvml,accsvmnl,accnav,accknn,accdes,accrf)
```
