```python
# important requriment library
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt
```

```python
def selectkbest(indep_x,dep_y,n):
    test=selectKBest(score_function=chi2,k=n)
    fit1=test.fit(indep_x,dep_y)
    selectk_features=fit1.transform(indep_x)
    return selectk_features
```

```python
def split_scalar(indep_x,dep_y):
    x_train,x_test,y_train,y_test=train_test_split(indep_x,dep_y, test_size=0.25,random_state=0)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.fit_transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def cm_prediction(classification,x_test):
    y_pred=classifier.predicr(x_test)
    # Making the Confusion Matrix
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    Accuracy=accuracy_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def logistic(x_train,y_train,x_test):
    # Fitting K-NN to the Training set
    from sklearn.linear_model import logistic_regression
    classifier =logistic_regression(random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def svm_linear(x_train,y_train,x_test):
    # Fitting K-NN to the Training set
    from sklearn.svm import svc
    classifer =SVC(kernel = 'linear',random_stae=0)
    classifer.fit(x_train,y_train)
    classifer,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifer,Accuraacy,report,x_test,y_test,cm
```

```python
def svm_NL(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='rbf', gamma='scale',random_state=0)
    classifier=classifier.fit(x_train,y_train)
    classifer,Accuracy,report,x_test,y_test,cm=cm_prediction(classifer,x_test)
    return classifier,Accuuracy,report,x_test,y_test,cm
```

```python
def Navie(x_train,y_train,x_test):
    # Fitting K-NN to the Training set
    from sklearn.Navie_nayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def knn(x_train,y_train,x_test):
    # Fitting K-NN to the Training set
    from sklearn.neighbors import KNeighborsClassifier
    classifier=KNeighborsClassifier(n_neighbors=7,weights='uniform',algorithm='auto',leaf_size=30,p=2,metric='minkowski')
    classifier.fit(x_train,y_train)
    classifier,Accuraacy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def DecisionTree(x_train,y_train,x_test):
    # Fitting K-NN to the Training set
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier(criterion='entropy', splitter='random',max_features='sqrt',random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def RandomForest(x_train,y_train,x_test):
     # Fitting K-NN to the Training set
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier(n_estimators = 10, criterion = 'entropy',max_features='sqrt', random_state = 0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm
```

```python
def Selectk_Classification(acclog,accsvml,accsvmNL,accNavie,accknn,accDecisionTree,accRandomForest):
    dataframe=pd.DataFrame(index=['ChiSquare'],columns=['logistic','svml','svmNL','Navie','knn','DecisionTree','RandomForest'])
    for number, index in enumerate(dataframe.index):
        dataframe['logistic'][index]=acclog[number]
        dataframe['svml'][index]=accsvml[number]
        dataframe['svmNL'][index]=accsvmNL[number]
        dataframe['Navie'][index]=accNavie[number]
        dataframe['knn'][index]=accknn[number]
        dataframe['DecisionTree']=accDecisionTree[number]
        dataframe['RandomForest']=accRandomForest[number]
        return dataframe
```
